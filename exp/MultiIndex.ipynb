# UnStack and Stack Data

```python
import pandas as pd
data = pd.read_csv('drinks.csv', index_col=['country', 'continent'])
```

# Pivot and Pivot_table Method

```python
data = pd.read_csv('drinks.csv')
```

```python
data.head()
```

```python
data.pivot(index='country', columns='continent', values=['spirit_servings'])
```

```python
data.pivot(index='country', columns='continent', values=['spirit_servings','beer_servings'])
```

```python
data.pivot_table(index='continent', values=['spirit_servings','beer_servings'], aggfunc='mean')
```

```python
data.pivot_table(index='country', values=['spirit_servings','beer_servings'], aggfunc='mean')
```
