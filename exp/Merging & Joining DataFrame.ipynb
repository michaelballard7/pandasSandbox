# Merging & Joining DataFrame

```python
import pandas as pd
```

```python
engg = pd.read_csv('data/Engineering - Student.csv')
science = pd.read_csv('data/Science - Student.csv')
exam = pd.read_csv('data/Test - Info.csv')
student_exam = pd.read_csv('data/Student - Exam - Info.csv')
```

```python
student_exam
```

```python
engg
```

# Concatenating two Data Frame

```python
pd.concat([engg, science])
```

```python
pd.concat([engg, science], axis=1)
```

```python
pd.concat([engg, science], ignore_index=True)
```

```python
pd.concat([engg, science], keys=['Engineering','Science'])
```

```python
engg.append(science, ignore_index=True)
```

```python
engg.merge(science, how='inner', on='student_id', suffixes=['_engg','_science'])
```

```python
student_exam.merge(engg, on='student_id')
```

```python
student_exam.merge(exam, on='exam_id')
```
